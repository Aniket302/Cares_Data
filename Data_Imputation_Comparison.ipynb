# Data Imputation using Mean

In this section, we impute the missing values in the ``RCRI score`` column and ``PreopEGFRMDRD`` column with the **mean** value of the respective columns.

Note that data imputation using Mean can be done for the numerical columns only.

```python
df_mean = df.copy()

df_mean['RCRI score'].fillna(df_mean['RCRI score'].mean(), inplace = True)
df_mean['PreopEGFRMDRD'].fillna(df_mean['PreopEGFRMDRD'].mean(), inplace = True)
```

```python
#Check that there are no more missing values in these two columns:
print(df_mean['RCRI score'].isna().sum())
print(df_mean['PreopEGFRMDRD'].isna().sum())
```

```python
df_mean.describe()
```

# Data Imputation using Median

In this section, we impute the missing values in the ``RCRI score`` column and ``PreopEGFRMDRD`` column with the **median** value of the respective columns.

Similar to Mean data imputation, this can be done for the numerical columns only.

```python
df_median = df.copy()

median_RCRI = df_median['RCRI score'].median()
df_median['RCRI score'].fillna(median_RCRI, inplace = True)

median_PreopEGFRMDRD = df_median['PreopEGFRMDRD'].median()
df_median['PreopEGFRMDRD'].fillna(median_PreopEGFRMDRD, inplace = True)
```

```python
df_median.describe()
```

Using Median data imputation, we observe that:

- The mean of **RCRI score** decreased (from 0.322 to 0.225) after replacing missing values in the ``RCRI score`` column with the median.
- The mean of **PreopEGFRMDRD** decreased slightly (from 96.426 to 96.423) after replacing missing values in the ``PreopEGFRMDRD`` column with the median.

The advantage of using Mean data imputation or Median data imputation is that both methods are easy to understand and apply.

However, it does not take into account the variability within the data.

# Linear Interpolation imputation

In this section, we impute the missing values in the ``RCRI score`` column and ``PreopEGFRMDRD`` column using linear interpolation, with Backward fill to handle missing values at the end of the DataFrame, and Forward fill imputations to handle missing values at the start of the DataFrame.

Linear interpolation imputation can be done for the numerical columns only.

```python
df_interpolation = df.copy()
```

```python
# Interpolation imputation for RCRI score and PreopEGFRMDRD
df_interpolation['RCRI score'].interpolate(method='linear', inplace=True)
df_interpolation['PreopEGFRMDRD'].interpolate(method='linear', inplace=True)

# Forward fill to handle missing values at the start
df_interpolation['RCRI score'].fillna(method='ffill', inplace=True)
df_interpolation['PreopEGFRMDRD'].fillna(method='ffill', inplace=True)

# Backward fill to handle missing values at the end
df_interpolation['RCRI score'].fillna(method='bfill', inplace=True)
df_interpolation['PreopEGFRMDRD'].fillna(method='bfill', inplace=True)
```

```python
#Check that there are no more missing values in these two columns:
print(df_interpolation['RCRI score'].isna().sum())
print(df_interpolation['PreopEGFRMDRD'].isna().sum())
```

```python
df_interpolation.describe()
```

Using linear interpolation, forward fill and backward fill to impute missing values, we observe that:

- The mean of the **RCRI score** decreased slightly from 0.322 to 0.316

- The mean of the **PreopEGFRMDRD** increased slightly from 96.426 to 97.788

# MICE imputation

In this section, since MICE imputation can be done on both continuous and categorical variables, we conduct MICE imputation on the following columns:

- Continuous variables: ``RCRI score``, ``PreopEGFRMDRD``
- Boolean (binary) variables: ``CVARCRICategory``, ``IHDRCRICategory``, ``CHFRCRICategory``, ``DMinsulinRCRICategory``, ``CreatinineRCRICategory``

```python
df_mice = df.copy()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Using MICE for imputation
mice_imputer = IterativeImputer()
columns = ['RCRI score', 'PreopEGFRMDRD','CVARCRICategory', 'IHDRCRICategory', 'CHFRCRICategory',
           'DMinsulinRCRICategory', 'CreatinineRCRICategory']
bool_columns = ['CVARCRICategory', 'IHDRCRICategory', 'CHFRCRICategory', 'DMinsulinRCRICategory', 'CreatinineRCRICategory']
imputed_data = mice_imputer.fit_transform(df_mice[columns])

# Convert imputed_data back to a DataFrame
imputed_df = pd.DataFrame(imputed_data, columns = columns)

# Convert the values in 'CVARCRICategory', 'IHDRCRICategory', 'CHFRCRICategory','DMinsulinRCRICategory', 
# 'CreatinineRCRICategory' columns back to Boolean data type
imputed_df[bool_columns] = imputed_df[bool_columns].round().astype(bool)
```

```python
# Replace NaN values with computed values
df_mice[columns] = imputed_df[columns]

imputed_df
```

```python
#Check that there are no more missing values in these columns:
print(df_mice['RCRI score'].isna().sum())
print(df_mice['PreopEGFRMDRD'].isna().sum())
print(df_mice['CVARCRICategory'].isna().sum())
print(df_mice['IHDRCRICategory'].isna().sum())
print(df_mice['CHFRCRICategory'].isna().sum())
print(df_mice['DMinsulinRCRICategory'].isna().sum())
print(df_mice['CreatinineRCRICategory'].isna().sum())
```

```python
df_mice.describe()
```

Using MICE imputation, we observe that:

- The mean of the **RCRI score** remained the same from 0.322 to 0.322

- The mean of the **PreopEGFRMDRD** increased slightly from 96.426 to 96.835

Comparing the number of TRUE and FALSE values in the other columns before and after MICE data imputation:

The advantage of MICE data imputation is that it can impute missing values for both continuous and categorical variables. However, it's more complex to implement compared to simpler methods like Mean or Median imputation.

# Regression imputation

In this section, we first use Mean data imputation to impute the missing values in the ``PreopEGFRMDRD`` column, then use Linear Regression to impute the missing values in the ``RCRI score`` column. 

Linear Regression imputation can be done for the numerical columns only.

```python
df_regression = df.copy()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer


# Step 1: Impute missing values in PreopEGFRMDRD using mean
imputer = SimpleImputer(strategy='mean')
df_regression['PreopEGFRMDRD'] = imputer.fit_transform(df_regression[['PreopEGFRMDRD']])

# Step 2: Regression imputation for RCRI score
regressor_rcri = LinearRegression()
non_na_data_rcri = df_regression.dropna(subset=['RCRI score', 'PreopEGFRMDRD'])
regressor_rcri.fit(non_na_data_rcri[['PreopEGFRMDRD']], non_na_data_rcri['RCRI score'])

missing_rcri = df_regression[df_regression['RCRI score'].isna()]
if not missing_rcri.empty:
    df_regression.loc[df_regression['RCRI score'].isna(), 'RCRI score'] = regressor_rcri.predict(missing_rcri[['PreopEGFRMDRD']])

# Check if there are still missing values in PreopEGFRMDRD after mean imputation
if df_regression['PreopEGFRMDRD'].isna().sum() > 0:
    # Regression imputation for PreopEGFRMDRD
    regressor_preop = LinearRegression()
    non_na_data_preop = df_regression.dropna(subset=['PreopEGFRMDRD', 'RCRI score'])
    regressor_preop.fit(non_na_data_preop[['RCRI score']], non_na_data_preop['PreopEGFRMDRD'])

    missing_preop = df_regression[df_regression['PreopEGFRMDRD'].isna()]
    if not missing_preop.empty:
        df_regression.loc[df_regression['PreopEGFRMDRD'].isna(), 'PreopEGFRMDRD'] = regressor_preop.predict(missing_preop[['RCRI score']])
```

```python
#Check that there are no more missing values in these two columns:
print(df_regression['RCRI score'].isna().sum())
print(df_regression['PreopEGFRMDRD'].isna().sum())
```

```python
df_regression.describe()
```

Using Regression imputation, we observe that:

- The mean of the **RCRI score** slightly decreased from 0.322 to 0.321

- The mean of the **PreopEGFRMDRD** remained the same from 96.426 to 96.426

Although Regression imputation uses the relationships between variables which may lead to more accurate imputations than simple methods like Mean or Median imputation, it assumes that the relationships between the variables is linear, which might not be actually true.

# Random Forest Imputation

```python
df_rf = df.copy()
```

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

# Step 1: Impute missing values in PreopEGFRMDRD using mean
imputer = SimpleImputer(strategy='mean')
df_rf['PreopEGFRMDRD'] = imputer.fit_transform(df_rf[['PreopEGFRMDRD']])

# Step 2: Random forest imputation for RCRI score
rf_imputer_rcri = RandomForestRegressor(n_estimators=100)
non_na_data_rcri = df_rf.dropna(subset=['RCRI score', 'PreopEGFRMDRD'])
rf_imputer_rcri.fit(non_na_data_rcri[['PreopEGFRMDRD']], non_na_data_rcri['RCRI score'])

missing_rcri = df_rf[df_rf['RCRI score'].isna()]
if not missing_rcri.empty:
    df_rf.loc[df_rf['RCRI score'].isna(), 'RCRI score'] = rf_imputer_rcri.predict(missing_rcri[['PreopEGFRMDRD']])

# Check if there are still missing values in PreopEGFRMDRD after mean imputation
if df_rf['PreopEGFRMDRD'].isna().sum() > 0:
    # Random forest imputation for PreopEGFRMDRD
    rf_imputer_preop = RandomForestRegressor(n_estimators=100)
    non_na_data_preop = df_rf.dropna(subset=['PreopEGFRMDRD', 'RCRI score'])
    rf_imputer_preop.fit(non_na_data_preop[['RCRI score']], non_na_data_preop['PreopEGFRMDRD'])

    missing_preop = df_rf[df_rf['PreopEGFRMDRD'].isna()]
    if not missing_preop.empty:
        df_rf.loc[df_rf['PreopEGFRMDRD'].isna(), 'PreopEGFRMDRD'] = rf_imputer_preop.predict(missing_preop[['RCRI score']])
```

```python
#Check that there are no more missing values in these two columns:
print(df_rf['RCRI score'].isna().sum())
print(df_rf['PreopEGFRMDRD'].isna().sum())
```
